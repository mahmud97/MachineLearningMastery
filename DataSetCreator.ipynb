```python
import pandas as pd
import os
import shutil
```

Create the data for positive samples

```python
file_path = "CovidChestxray/metadata.csv"
images_path = "CovidChestxray/images"
```

```python
df = pd.read_csv(file_path)
print(df.shape)
```

```python
df.head(20)
```

```python
target_dir = "Dataset/Train/Covid"

if not os.path.exists(target_dir):
    os.mkdir(target_dir)
```

```python
# we need only front view 
# sampling images from covid postiive chest xray
cnt = 0 

for (i,row) in df.iterrows():
    if row["finding"]=="Pneumonia/Viral/COVID-19" and row["view"]=="PA":
        filename = row["filename"]
        image_path = os.path.join(images_path,filename)
        image_copy_path = os.path.join(target_dir,filename)
        shutil.copy2(image_path,image_copy_path)
        #print("Moving image",cnt)
        cnt+=1
    
print("total no of covid19 positive images from chest-x-ray-github are: ",cnt)
```

```python
# sampling images from kaggle normal
import random
kaggle_file_path = "KaggleData/train/NORMAL"
target_normal_dir = "Dataset/Train/Normal"
if not os.path.exists(target_normal_dir):
    os.mkdir(target_normal_dir)

```

```python
image_names = os.listdir(kaggle_file_path)
image_names
```

```python
random.shuffle(image_names)
```

```python
cnt = 0 
for i in range(196):
    image_name = image_names[i]
    image_path = os.path.join(kaggle_file_path,image_name)
    target_path = os.path.join(target_normal_dir,image_name)
    shutil.copy2(image_path,target_path)
    cnt+=1
    #print("copy image to directory", image_name)
print("total number of normal images ",cnt)
    
```

```python
target_test_dir_covid = "Dataset/Test/Covid"
target_test_dir_normal = "Dataset/Test/Normal"

if not os.path.exists(target_test_dir_covid and target_test_dir_normal ):
    os.mkdir(target_test_dir_covid)
    os.mkdir(target_test_dir_normal)
    
```

```python
image_normal_from_train = os.listdir(target_normal_dir)
image_normal_from_train
```

```python
random.shuffle(image_normal_from_train)
```

```python
cnt = 0 
for i in range(40):
    image_name =image_normal_from_train[i]
    image_path = os.path.join(target_normal_dir,image_name)
    target_path = os.path.join(target_test_dir_normal,image_name)
    shutil.copy2(image_path,target_path)
    cnt+=1
    #print("copy image to directory", image_name)
print("total number of normal images copying to test normal directory",cnt)
    
```

```python
# after transfering 40 images from Train normal to Test normal now cheking no of files in train normal directory 
file_count = sum(len(files) for _, _, files in os.walk(r'Dataset/Train/Normal'))
print(file_count)
```

```python
# now copying covid data from train directory to test directory
image_covid_from_train = os.listdir(target_dir)
image_covid_from_train
```

```python
random.shuffle(image_covid_from_train)
```

```python
cnt = 0 
for i in range(40):
    image_name =image_covid_from_train[i]
    image_path = os.path.join(target_dir,image_name)
    target_path = os.path.join(target_test_dir_covid,image_name)
    shutil.copy2(image_path,target_path)
    cnt+=1
    #print("copy image to directory", image_name)
print("total number of covid images copying to test normal directory",cnt)
```
